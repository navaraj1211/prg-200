# Temperature Logger (math module)

```python
import math

station_name = "Kathmandu Weather Station"

temperatures = [18.4, 22.1, 15.7, 29.3, 11.8, 25.6, 19.2]


def get_average(temps):
    total = sum(temps)
    average = total / len(temps)
    return average


def get_deviation(temps):
    mean = get_average(temps)   # local variable

    total = 0

    for temp in temps:
        total = total + (temp - mean) ** 2

    variance = total / len(temps)
    deviation = math.sqrt(variance)

    return deviation


def get_summary(temps):
    print(station_name)
    print("Minimum Temperature:", min(temps))
    print("Maximum Temperature:", max(temps))
    print("Average Temperature:", round(get_average(temps), 2))
    print("Standard Deviation:", round(get_deviation(temps), 2))


get_summary(temperatures)

```

# Bill Splitter (random module)

```python
import random

random.seed(42)

friends = ["Ramesh", "Sunita", "Bikash", "Anjali", "Dipak"]

total_bill = 3750


def split_bill(friends, total):
    share = total / len(friends)
    return share


def pick_lucky(friends):
    lucky_person = random.choice(friends)
    return lucky_person


def final_summary(friends, total):
    share = split_bill(friends, total)
    lucky = pick_lucky(friends)

    print("Bill Split Summary")
    print("------------------")

    for friend in friends:
        print(f"{friend} pays NPR {share}")

    lucky_total = share + 50

    print()
    print(f"Lucky Person: {lucky}")
    print(f"{lucky} pays extra NPR 50")
    print(f"{lucky}'s total payment: NPR {lucky_total}")

```

# Exam Scheduler (datetime module) 

```python
from datetime import datetime, timedelta

college_name = "Bhaktapur Multiple Campus"

start_date = "2025-05-01"

exams = [
    ("Python Programming", 0),
    ("Data Structures", 3),
    ("Database Systems", 6),
    ("Computer Networks", 10),
    ("Mathematics", 14)
]


def parse_date(date_str):
    date_object = datetime.strptime(date_str, "%Y-%m-%d")
    return date_object


def get_exam_date(start_str, days):
    start = parse_date(start_str)
    exam_date = start + timedelta(days=days)
    return exam_date.strftime("%Y-%m-%d")


def print_schedule(start_str, exams):
    print(college_name)
    print("Exam Schedule")
    print("----------------")

    for subject, days in exams:
        exam_date = get_exam_date(start_str, days)
        print(f"{subject}: {exam_date}")

```

# Shopping Discount Own Module

```python
import discount

products = [
    ("Laptop", 85000, 10),
    ("Headphones", 4500, 15),
    ("Phone Case", 800, 5),
    ("USB Cable", 600, 0)
]

print("Tax Rate:", discount.TAX_RATE)
print("Product Price Summary")
print("---------------------")

for name, price, discount_percent in products:
    final = discount.final_price(price, discount_percent)

    print("Product:", name)
    print("Original Price: NPR", price)
    print("Discount:", discount_percent, "%")
    print("Final Price: NPR", round(final, 2))
    print()
```
